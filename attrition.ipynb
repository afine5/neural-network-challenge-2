## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
```

```python
# Create a list of at least 10 column names to use as X data
col_12_list = ['Age',
              'YearsInCurrentRole',
              'DistanceFromHome',
              'Education',
              'EnvironmentSatisfaction',
              'HourlyRate',
              'NumCompaniesWorked',
              'JobLevel',
              'PercentSalaryHike',
              'YearsAtCompany',
              'TotalWorkingYears',
              'RelationshipSatisfaction']

# Create X_df using your selected columns
X_df = attrition_df[col_12_list]

# Show the data types for X_df
# X_df.dtypes
X_df.info()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
# Convert X data to numeric data types
for column in X_df.columns:
    if X_df[column].dtype == 'object':
        try:
            X_df[column] = pd.to_numeric(X_df[column])
            print(f"Column '{column}' converted to numeric.")
        except ValueError:
            print(f"Column '{column}' could not be converted to numeric.")
    else:
        print(f"Column '{column}' is already numeric.")
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Transform the testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Scale the training and testing data
```

```python
# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
ohe.fit(np.array(y_train['Department']).reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
y_dept_encoded_train = ohe.transform(np.array(y_train['Department']).reshape(-1,1))
y_dept_encoded_test = ohe.transform(np.array(y_test['Department']).reshape(-1,1))
```

```python
# Create a OneHotEncoder for the Attrition column
ohe_attrition = ohe.fit(np.array(y_train['Attrition']).reshape(-1,1))

# Fit the encoder to the training data
ohe_attrition.fit(np.array(y_train['Attrition']).reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
y_attrition_encoded_train = ohe_attrition.transform(np.array(y_train['Attrition']).reshape(-1,1))
y_attrition_encoded_test = ohe_attrition.transform(np.array(y_test['Attrition']).reshape(-1,1))
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_features = X_train.shape[1]

# Create the input layer
input = layers.Input(shape=(input_features,))

# Create at least two shared layers
shared_layer1 = layers.Dense(128, activation='relu')(input)
shared_layer2 = layers.Dense(64, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer


# Create the hidden layer
dept_layer = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
dept_layer_output = layers.Dense(y_dept_encoded_train.shape[1], activation='softmax', name='department_output')(dept_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_layer = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
attrition_layer_output = layers.Dense(y_attrition_encoded_train.shape[1], activation='sigmoid', name='attrition_output')(attrition_layer)
```

```python
# Create the model
model = Model(inputs=input, outputs=[dept_layer_output, attrition_layer_output])

# Compile the model
model.compile(optimizer='adam', loss={'department_output': 'categorical_crossentropy',
                                      'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy',
                                      'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled,
          {'department_output': y_dept_encoded_train,
           'attrition_output': y_attrition_encoded_train},
          epochs=100)
```

```python
# Evaluate the model with the testing data
model_result = model.evaluate(X_test_scaled,
               {'department_output': y_dept_encoded_test,
                'attrition_output': y_attrition_encoded_test})
```

```python
# Print the accuracy for both department and attrition
print(f'department accuracy = {model_result[1]}')
print(f'attrition accuracy = {model_result[2]}')
```
